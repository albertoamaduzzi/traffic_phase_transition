```python
import paramiko 
import os
def upload_files(local_paths, remote_paths, hostname, username, password=None, key_filename=None):
    # Create an SSH client
    client = paramiko.SSHClient()
    client.set_missing_host_key_policy(paramiko.AutoAddPolicy())

    try:
        # Connect to the remote machine
        if password:
            client.connect(hostname, username=username, password=password)
        elif key_filename:
            client.connect(hostname, username=username, key_filename=key_filename)
        else:
            raise ValueError("Either password or key_filename must be provided.")

        # Create a transport object
        transport = paramiko.Transport((hostname, 22))
        transport.connect(username=username, password=password, pkey=None)

        # Create an SFTP client
        sftp = paramiko.SFTPClient.from_transport(transport)

        # Upload each file
        for local_path, remote_path in zip(local_paths, remote_paths):
            sftp.put(local_path, remote_path)
            print(f"File {local_path} uploaded to {remote_path}")

    finally:
        # Close the connections
        if sftp:
            sftp.close()
        if transport:
            transport.close()
        client.close()

# Example usage
local_files = ["file1", "file2", "file3"]
remote_files = ["/path/to/remote/file1", "/path/to/remote/file2", "/path/to/remote/file3"]

upload_files(local_files, remote_files, "remote_hostname", "remote_username", password="remote_password")
```
